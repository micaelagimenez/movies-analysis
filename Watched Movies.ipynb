```python
#Importing necessary packages
import pandas as pd
import datetime as dt
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## Taking a look at the data
```

```python
#Examining data
df = pd.read_csv('C:/Users/micae/Documents/dataset/letterboxd/watched.csv')
df.head()
list(df.columns)

diary = pd.read_csv('C:/Users/micae/Documents/dataset/letterboxd/diary.csv', encoding='latin-1')
diary.head()
list(diary.columns)
```

```python
## Visualizing the data
```

```python
### Movies watched in 2018 and 2019
```

```python
#Movie count 2018-2019
df['Date'] = pd.to_datetime(df['Date']) 
df['Years'] = df['Date'].dt.year

movies_by_year = df['Name'].groupby(df['Years']).count().reset_index()
movies_by_year.head()
two_years = movies_by_year.loc[[3, 4]]

sns.set(style='white')
ax = sns.barplot(x='Years', y='Name', data=two_years)
plt.show()
```

```python
### Movies watched by year they were made
```

```python
#Movies watched by year they were made
df['Year']
movies_year = df['Name'].groupby(df['Year']).count().reset_index()
movies_year.head()

plt.figure(figsize=(15,10))
sns.set(style='white')
ax = sns.barplot(x='Year', y='Name', data=movies_year)
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.title('Watched movies by years in which they were made')
plt.show()
```

```python
### Movies watched by rating
```

```python
#Movies by rating count
movies_rating = diary['Name'].groupby(diary['Rating']).count().reset_index()
movies_rating

plt.figure(figsize=(15,10))
ax = sns.barplot(x='Rating', y='Name', data=movies_rating, palette='gist_earth_r')
plt.title('Movies by rating')
plt.show()
```
